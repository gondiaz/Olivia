```python
histofile_old = './histograms_pmaps.h5'
histofile_new = './histograms_pmaps_new.h5'
```

```python
histo_manager_old = get_histograms_from_file(histofile_old)
histo_manager_new = get_histograms_from_file(histofile_new)
```

```python
histonames = histo_manager_old.histos

for i, histoname in enumerate(histonames):
    
    if len(histo_manager_old[histoname].bins)==2:
        
        fig, axes = plt.subplots(1, 2, figsize=(8 * 2, 4 * 1))
        
        plot_histogram (histo_manager_old  [histoname], ax=axes[0], plot_errors=False, normed=True)
        plot_histogram (histo_manager_new  [histoname], ax=axes[1], plot_errors=False, normed=True)
```
